### 2) 항아리 모양 그래프 그리기

```python
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '신도림' in row[0]:
        for i in range (0, 101):
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))

f.reverse()

plt.title('신도림 지역의 인구 구조')
plt.rc('font', family='Gulim')
plt.rcParams['axes.unicode_minus'] = False # 한국어 깨질때도 쓸 수 있음
plt.barh(range(101),m, label="남성")
plt.barh(range(101),f, label="여성")
plt.show()
```

```python
f = open('gender.csv')
data = csv.reader(f)
result = []

name = input('궁금한 동네를 입력해주세요:')

for row in data:
    if name in row[0]:
        for i in range(3, 104):
            result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
        break
        
plt.bar(range(101), result)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

name = input('궁금한 동네를 입력해주세요:')

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(int(i.replace(',','')))
        for i in row[106: ]:
            f.append(int(i.replace(',','')))
        break
        
plt.scatter(m,f, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g') # 'g'추세선 추가
plt.show()
```

마지막으로 연령별 남성과 여성 인구수를 합친 값을 size 리스트에 넣어 점의 크기를 표현한다. 이 때, math 라이브러리에 있는 제곱근 함수 sqrt()를 사용하여 점의 크기를 적절히 조절하고 x, y축의 이름을 xlabe(), ylabe() 함수로 정해준다.

```python
import matplotlib.pyplot as plt
import csv
import math

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
size = [] ## 

name = input('궁금한 동네를 입력해주세요: ')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
            size.append(math.sqrt(int(row[i].replace(',',''))+int(row[i+103].replace(',','')))) ##
        break
# 사이즈 오류 해결 --> replace 쓰면댄다 ㅠ_ㅠ 
        
plt.style.use('ggplot') ##
plt.rc('font', family='Gulim') ##
plt.rcParams['axes.unicode_minus'] = False ##
plt.figure(figsize=(10,5), dpi=300) ##
plt.title(name+' 지역의 성별 인구 그래프')
plt.scatter(m, f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g') # 'g'추세선 추가
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
