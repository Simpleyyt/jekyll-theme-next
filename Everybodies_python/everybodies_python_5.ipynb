### 2) pandas란? 
pandas는 panel datas(패널 자료)의 약자이다. numpy를 기반으로 만들어졌으며 데이터 분석을 위한 효율적인 데이터 구조를 제공하는데 1차원 형태의 데이터 구조를 <b>Series</b>라고 부르고 2차원 배열 형태의 데이터 구조를 <b>Data Frame</b>이라고 부른다. 조금 전에 올림픽 표를 저장했던 df라는 변수도 data frame을 뜻하는 변수이다. 

먼저 날짜 형태로 된 8개의 인덱스를 간단히 만든다. 그리고 numpy 라이브러리를 활용하여 8행 3열로 구성된 랜덤 데이터를 생성한 후, 인덱스와 컬럼 이름을 정하여 데이터 프레임으로 만든다.

```python
import pandas as pd
index = pd.date_range('1/1/2000', periods=8)
print(index)

import numpy as np
df = pd.DataFrame(np.random.rand(8, 3), index=index, columns=list('ABC'))
df
```

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
print(index)

df = pd.DataFrame(np.random.rand(8, 3), index=index, columns=list('ABC'))
df['B']>0.4 # mask 기능 사용
```

```python
df2 = df[df['B'] > 0.4] # mask 기능을 사용해서 일부분만 데이터 프레임으로 저장
df2
```

```python
df2.T # .T는 행과 열을 바꾸는 함수이다. 
```

데이터 프레임은 axis로 행 우선 계산과 열 우선 계산을 지정 할 수 있는데, axis=0이면 행↓방향 계산, axis=1이면 열→ 방향 계산이 가능하다. pandas는 기본적으로 <b>행 방향 계산을 우선시하지만</b> 열 방향으로 계산하려면 axis=1로 바꾸어 주면 된다.

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
print(index)

df = pd.DataFrame(np.random.rand(8, 3), index=index, columns=list('ABC'))
df['D'] = df['A'] / df['B']
df
```

다음 예제는 df 데이터 프레임에서 행 우선으로 합을 구하고 그 결과를 E라는 열을 새로 생성해서 저장하는 코드이다. head()함수는 많은 데이터 중 처음 5개의 데이터만 확인하고 싶을 때 사용하는 함수로 많은 양의 데이터의 형태를 확인 할 때 유용하다.

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
print(index)

df = pd.DataFrame(np.random.rand(8, 3), index=index, columns=list('ABC'))
df['D'] = df['A'] / df['B']
df['E'] = np.sum(df, axis=1)
df.head()
```

다음은 전체 데이터에서 열 우선 계산을 하는 방법이다. 아까는 1개의 열을 1개의 열로 나누는 계산이었기 때문에 산술연산자를 사용할 수 있었지만 여러 개의 열에 대한 계산을 하려면 pandas 라이브러리를 이용해야 한다. 예로 전체 데이터를 A열의 값으로 뺄 때는 sub() 함수를 사용한다.

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
print(index)

df = pd.DataFrame(np.random.rand(8, 3), index=index, columns=list('ABC'))
df['D'] = df['A'] / df['B']
df['E'] = np.sum(df, axis=1)
df = df.sub(df['A'], axis=0)
df.to_csv('test.csv')
df.head()
```

### 3) pandas로 인구 구조 분석하기
1. 데이터를 읽어온다.<br>
  1) 전체 데이터를 총 인구수로 나누어 비율로 변환한다.<br>
  2) 총인구수와 연령구간인구수를 삭제한다.<br>
2. 궁금한 지역의 이름을 입력받는다.<br>
3. 궁금한 지역의 인구 구조를 저장한다.<br>
4. 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역을 찾는다.<br>
  1) 전국의 모든 지역 중 한 곳(B)을 선택한다.<br>
  2) 궁금한 지역 A의 0세 인구 비율에서 B의 0세 인구 비율을 뺀다.<br>
  3) 2를 100세 이상 인구수에 해당하는 값까지 반복한 후 차이의 제곱을 모두 더한다.<br>
  4) 전국의 모든 지역에 대해 반복하며 그 차이가 가장 작은 지역을 찾는다.<br>
5. 가장 비슷한 곳의 인구 구조와 궁금한 지역의 인구 구조를 시각화한다. <br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. 먼저 데이터를 읽어온다.
df = pd.read_csv('201902_201902_montly.csv', encoding='cp949', index_col=0)

# 2. 그리고 연령별 인구 숫자를 비율로 바꿔주고 총 인구수와 연령구간인구수 데이터를 삭제한다. 특정 열을 삭제 할 때는 del을 사용하면 된다.
df = df.div(df['총인구수'], axis=0) # 전체 데이터를 총인구수로 나눠서 비율로 변환
del df['총인구수'], df['연령구간인구수']

# 3. 궁금한 지역의 이름을 입력받고 해당 지역의 인구 구조를 저장한다.

name = input('원하는 지역의 이름을 입력해주세요: ')
a = df.index.str.contains(name)
df2 = df[a]

df.head()
df2.head()
```

```python
plt.rc('font', family ='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
df2.T.plot()
plt.show()
```

```python
# 4. 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역을 시각화한다.

# 궁금한 지역 A의 인구 비율에서 B의 인구 비율을 뺀다.
x = df.sub(df2.iloc[0], axis=1) 

# a의 인구 비율에서 B의 인구 비율을 뺀 값의 제곱 값을 모두 더한다.
y = np.power(x,2)
z = y.sum(axis=1)

# 그 차이가 가장 작은 지역 5곳을 찾는다.
i = z.sort_values().index[:5]

# 결과를 꺾은 선 그래프로 보여준다.
df.loc[i].T.plot()
plt.show()
```

혹은 다음처럼 한 줄로 나타낼 수도 있다.

```python
df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_values().index[:5]].T.plot()
```
