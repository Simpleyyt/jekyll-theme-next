<b>● 마스크(mask) 기능</b><br>
　1) 마스크는 어떤 조건에 부합하는 데이터만 선별적으로 저장하기 위한 기능이다. 아래와 같이 -5부터 5까지의 숫자가 저장된 배열이 있을 때, numpy에서는 배열 전체에 조건을 적용해서 2)번 처럼 표현할 수 있다. 
<br> 　2) 이 때, 배열에서 조건에 부합하는 데이터는 True, 그렇지 않은 데이터는 False인 마스크가 생성이 된다.
<br> 　3) 이 마스크를 배열에 적용하면 조건에 부합하는(True) 데이터만 남게된다. 
<br> 　4) 또한 마스크를 변수에 넣어서 사용할 수도 있다.
<br> 　5) 여러개의 마스크를 연결해서 사용하는 것 또한 가능하다.

```python
a = np.arange(-5, 5)
print(a) 
print(a<0) # 2)
print(a[a<0]) # 3)
print('')
# 4)
mask1 = abs(a) > 3
print(a[mask1])
print('')
# 5)
mask2 = abs(a) % 2==0
print(a[mask1+mask2])
print(a[mask1*mask2])
```

3) 궁금한 지역의 인구 구조를 저장한다.

```python
import csv
f = open('201902_201902_montly.csv')
data = csv.reader(f)
next(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요: ')
home = []

for row in data:
    if name in row[0]:
        for i in row[3: ]:
            home.append(int(i))

print(home)
```

위 코드를 numpy 라이브러리를 사용해 수정한다. 자동으로 int 형으로 반환되어 쉼표 없이 저장된다.

```python
import numpy as np
import csv

f = open('201902_201902_montly.csv')
data = csv.reader(f)
next(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요: ')

for row in data:
    if name in row[0]:
        for i in row[3: ]:
            home = np.array(row[3:], dtype=int)

print(home)
```

<b>Q1. 서로 다른 인구를 가진 두 지역을 어떻게 비교할 수 있을까?</b><br>
　2번의 알고리즘으론 비교할 수 없다. 수정이 필요하다!<br>
　(1) 전국의 모든 지역 중 한 곳(B)을 선택한다.<br>
　(2) 궁금한 지역 A의 0세 '인구 비율'에서 B의 0세 '인구 비율'을 뺀다. <br>
　(3) 2를 100세 이상 인구수에 해당하는 값까지 반복한 후 각각의 차이를 모두 더한다.<br>
　(4) 전국의 모든 지역에 대해 반복하며 그 차이가 가장 작은 지역을 찾는다.<br>

비율을 구하려면 그 지역의 전체 인구수로 연령대별 인구수를 나누면 된다. 데이터에서 2번 인덱스에 전체 인구수가 저장되어 있으므로 아래와 같이 코드를 수정한다. 다음은 화촌면 지역과의 비교를 나타낸 그래프를 추가하여 그렸다.

```python
import numpy as np
import csv

f = open('201902_201902_montly.csv')
data = csv.reader(f)
next(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요: ')
name2 = '화촌면'

for row in data:
    if name in row[0]:
        for i in row[3: ]:
            home = np.array(row[3:], dtype=int) / int(row[2].replace(',',''))
            
    if name2 in row[0]:
        for i in row[3: ]:
            home2 = np.array(row[3:], dtype=int) / int(row[2].replace(',',''))


#print(home)

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + '지역과 ' + name2 + ' 지역의 인구 구조')
plt.plot(home)
plt.plot(home2)
plt.show()
```

<b>Q2. 한 번 불러온 데이터를 여러 번 사용하려면 어떻게 해야 할까?</b>
data = list(data)를 추가한다. 

```python
import numpy as np
import csv
import matplotlib.pyplot as plt

f = open('201902_201902_montly.csv')
data = csv.reader(f)
next(data)
data = list(data) # 추가된 코드

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요: ')
mn = 1

result_name = ''
result = 0

for row in data:
    if name in row[0]:
        home = np.array(row[3:], dtype=int) / int(row[2])

for row in data:
    away = np.array(row[3: ], dtype=int) / int(row[2])
    s = np.sum(home - away) # home에 저장된 값과 away에 저장된 값의 차이의 합을 계산한다.
    if s < mn:
        mn = s
        result_name = row[0]
        result = away
        
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + '지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(away, label=result_name)
plt.legend()
plt.show()
## 이상함 책에서는 충청남도 논산시 연산면 나오는데 ㅇ_ㅇ? 
```

그러나 이 알고리즘에는 이상한 부분이 있다. 데이터 값이 음수인 부분이 채택되어 이상한 결과가 나오게 된다.

```python
import numpy as np
import csv
import matplotlib.pyplot as plt

f = open('201902_201902_montly.csv')
data = csv.reader(f)
next(data)
data = list(data) # 추가된 코드

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요: ')
mn = 1

result_name = ''
result = 0

for row in data:
    if name in row[0]:
        home = np.array(row[3:], dtype=int) / int(row[2])

for row in data:
    away = np.array(row[3: ], dtype=int) / int(row[2])
    s = np.sum((home - away)**2) # 수정된 알고리즘
    if s < mn:
        mn = s
        result_name = row[0]
        result = away
        
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + '지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(away, label=result_name)
plt.legend()
plt.show()
```

그러나 같은 지역이 나오므로 해당 지역을 배제하고 계산해본다. if s < mn and name not in row[0] : <br>

```python
import numpy as np
import csv
import matplotlib.pyplot as plt

f = open('201902_201902_montly.csv')
data = csv.reader(f)
next(data)
data = list(data) # 추가된 코드

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요: ')
mn = 1

result_name = ''
result = 0

for row in data :
    if name in row[0] :
        home = np.array(row[3:], dtype =int) /int(row[2])

for row in data :
    away = np.array(row[3:], dtype =int) /int(row[2])
    s = np.sum((home - away) **2)
    if s < mn and name not in row[0] :
        mn = s
        result_name = row[0]
        result = away
        
plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi=300)            
plt.rc('font', family ='Malgun Gothic')
plt.title(name +' 지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label = name, c='blue')
plt.plot(result, label = result_name, c='red')
plt.legend()
plt.show()
```
